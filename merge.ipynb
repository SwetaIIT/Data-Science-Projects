```python
import pprint 
import json
import glob
import os
import pandas as pd
```

```python
files = glob.glob("JSON_tasks_snapshot/**/*.json", recursive=True) 

tasks = {}
for taskFile in files:
  with open(taskFile, encoding="utf-8") as f:
    raw = json.loads(f.read())
  taskId = raw["_id"]["$oid"]

  if (taskId not in tasks.keys()):
    tasks[taskId] = {
    }
  for sentIndex, sentData in enumerate(raw["sents"]):
    sentId = sentData["sentId"]
    if (sentId not in tasks[taskId].keys()):
      tasks[taskId][sentId] = { "words": []}
    tasks[taskId][sentId]["words"] = sentData["words"]
    tasks[taskId][sentId]["text"] = sentData["text"]

tasks["5fc4c3a06c71e4f6c7030512"]["train-s998"]
```

```python
files = glob.glob("JSON_annotators/**/*.json", recursive=True) 
usernames = ["ex_shr_ann01", "ex_aan_ann02", "ex_sar_ann03", "ex_ars_ann04"]

table = {
    "annotator": [],
    "task": [],
    "wordRT": [],
    "wordTag": [],
    "sentRT": [],
    "sentId": [],
    "word": [],
    "text": []
  }
for taskFile in files:
  with open(taskFile, encoding="utf-8") as f:
    raw = json.loads(f.read())
  taskId = taskFile.split("/")[1]

  for sentIndex, sentData in enumerate(raw["data"]):
    for wordIndex, wordData in enumerate(sentData["wordRTs"]):
      table["annotator"].append(raw["username"])
      table["task"].append(taskId)
      table["sentId"].append(sentData["sentId"])
      table["sentRT"].append(sentData["sentRT"])
      table["wordRT"].append(sentData["wordRTs"][wordIndex])
      table["wordTag"].append(sentData["wordTags"][wordIndex])
      sentId = sentData["sentId"]
      table["word"].append(tasks[taskId][sentId]["words"][wordIndex])
      table["text"].append(tasks[taskId][sentId]["text"])

df = pd.DataFrame(table)
df

    
    
```
